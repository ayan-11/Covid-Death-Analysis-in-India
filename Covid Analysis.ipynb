```python
y, x = dataset['Current Status'], dataset['State code']

plt.figure(figsize=(15, 6))
plt.bar(x, y, color='lightblue')

plt.xlabel('State code')  # Add appropriate labels for the x-axis and y-axis
plt.ylabel('Current Status')
plt.title('Bar Plot of Current Status by State Code')  # Add a title for your plot
plt.show()
```

```python
# So we can see that different state conditions of the patients in different states of India.
```

```python
y, x = dataset['Date Announced'], dataset['Num Cases']

plt.figure(figsize=(15, 6))
plt.plot(x, y, color='lightblue', marker='o', linestyle='-')  # Use 'o' for markers and '-' for lines

plt.xlabel('Num Cases')
plt.ylabel('Date Announced')
plt.title('Line Graph of Num Cases by Date Announced')
plt.show()
```
